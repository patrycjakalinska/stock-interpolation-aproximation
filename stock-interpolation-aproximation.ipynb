```python
import pandas as pd
import numpy as np
import random as rnd
import matplotlib.pyplot as plt
import scipy as sc
from scipy import interpolate
from scipy.interpolate import Rbf
from scipy.fft import fft
from scipy.optimize import curve_fit
```

```python
def interpolation(x, y, xx):

  n = len(x)
  sum = 0
  for i in range(n):
    product = y[i]
    for j in range(n):
      if i!= j:
        product = product*(xx - x[j])/(x[i] - x[j])
    sum = sum + product
  return sum
  
```

```python
df1 = pd.read_csv(r'peo_m.csv', sep=",")
df1['Data'] = pd.to_datetime(df1.Data, format='%Y.%m.%d')
data1 = df1['Data']
value1 = np.array(df1['Najwyzszy'])
```

```python
# interpolacja Lagrange 1
x1 = np.linspace(1, df1.shape[0], df1.shape[0])
xx = np.linspace(1, df1.shape[0], df1.shape[0])
yy = [interpolation(x1, value1, i) for i in xx]

plt.figure(figsize=(12,10))
plt.grid(True, which='both')
plt.axvline(x=0, color = 'k')
plt.scatter(data1, value1)
plt.plot(data1,yy)
plt.ylabel('Value')
plt.xlabel('Date')
plt.title('Highest - PEO')
plt.xlim([data1[0],data1[df1.shape[0]-1]])
fig = plt.figure()

# interpolacja Scipy interp1d

f1 = interpolate.interp1d(x1, value1) 

plt.figure(figsize=(12,10))
plt.grid(True, which='both')
plt.axvline(x=0, color = 'k')
ynew1 = f1(x1)
plt.plot(data1,ynew1)
plt.scatter(data1, value1)
plt.ylabel('Value')
plt.xlabel('Date')
plt.title('Highest - PEO | INTERP1D')
plt.xlim([data1[0],data1[df1.shape[0]-1]])
fig = plt.figure()

# interpolacja Scipy Spline
temp = interpolate.splrep(x1, value1, s=0)
ynew2 = interpolate.splev(x1, temp, der=0)

plt.figure(figsize=(12,10))
plt.grid(True, which='both')
plt.axvline(x=0, color = 'k')
plt.plot(data1, ynew2)
plt.scatter(data1,value1)
plt.ylabel('Value')
plt.xlabel('Date')
plt.title('Highest - PEO | SPLINE')
plt.xlim([data1[0],data1[df1.shape[0]-1]])
fig = plt.figure()

# interpolacja Scipy Radial basis 
rbf = Rbf(x1, value1)
fi = rbf(x1)

plt.figure(figsize=(12,10))
plt.grid(True, which='both')
plt.axvline(x=0, color = 'k')
plt.plot(data1, fi)
plt.scatter(data1, value1)
plt.ylabel('Value')
plt.xlabel('Date')
plt.title('Highest - PEO | RADIAL BASIS')
plt.xlim([data1[0],data1[df1.shape[0]-1]])
fig = plt.figure()
```

```python
def TransformataFouriera(t, y):
  N = len(t)
  dt = t[1] - t[0]
  yf = 2.0 / N *np.abs(fft(y)[0:N//2])
  xf = np.fft.fftfreq(N, d=dt)[0:N //2]
  return xf, yf 
```

```python
# FFT PEO

xf, yf = TransformataFouriera(x1, value1)
plt.plot(xf, yf)
plt.grid()
plt.show()
```

```python
# FFT BET

xf, yf = TransformataFouriera(x1, value2)
plt.plot(xf, yf)
plt.grid()
plt.show()
```

```python
# FFT DJC

xf, yf = TransformataFouriera(x1, value3)
plt.plot(xf, yf)
plt.grid()
plt.show()
```
